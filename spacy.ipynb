```python
import spacy
import PyPDF2

# Lade deutsches Sprachmodell
nlp = spacy.load("de_core_news_md")

# Liste typischer Begriffe
KEYWORDS = [
    "nachhaltig", "klimaneutral", "umwelt", "klimaschutz", "verantwortung",
    "ressourcenschonend", "emission", "recyceln", "grün", "CO2", "klimafreundlich"
]

def pdf_zu_text(pfad):
    text = ""
    with open(pfad, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        for seite in reader.pages:
            seiten_text = seite.extract_text()
            if seiten_text:
                text += seiten_text + "\n"
    return text

def finde_nachhaltigkeitssaetze(text):
    doc = nlp(text)
    saetze = []
    for satz in doc.sents:
        if any(keyword in satz.text.lower() for keyword in KEYWORDS):
            saetze.append(satz.text.strip())
    return saetze

if __name__ == "__main__":
    pfad = "2023_Volkswagen_Group_Nachhaltigkeitsbericht.pdf"
    text = pdf_zu_text(pfad)
    versprechen = finde_nachhaltigkeitssaetze(text)

    if versprechen:
        print("\n--- Gefundene Nachhaltigkeitsversprechen (spaCy) ---\n")
        for i, satz in enumerate(versprechen, 1):
            print(f"{i}. {satz}")
    else:
        print("Keine Nachhaltigkeitsversprechen gefunden.")
```
