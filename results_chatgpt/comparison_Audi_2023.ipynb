### **Hintergründe & NGO-Bewertungen**

#### **1. Umweltmanagement & Kreislaufwirtschaft**

* Amnesty International kritisiert die Autoindustrie für unzureichende menschrechtliche Sorgfalt, inklusive Themen wie Cobalt‑ und Aluminiumversorgung – was Rückschlüsse auf Materialtransparenz zulässt ([Amnesty International][1]). Sustainalytics bewertet VW inkl. Audi mit soliden ESG‑Programmen, jedoch verbleibt mittleres Risiko auf Grund der generellen Branchenherausforderungen.

#### **2. SDG‑Umsetzung & robuster Infrastruktur**

* HRW und ECCHR dokumentieren Bedenken bezüglich Aluminium‑Lieferketten aus Xinjiang, verwehren vertrauenswürdige Audits und fordern bessere Mapping-Mechanismen. VW/Audi könnten indirekt betroffen sein, da Aluminium essentiell für Elektroautos ist ([ecchr.eu][3]).
* Amnesty-Ranking „Recharge for Rights“ zeigt, dass selbst Branchenführer nur moderat bei Menschenrechts-Due-Diligence abschneiden ([Amnesty International][1]).

#### **3. Soziale Verantwortung**

* ECCHR hat eine Beschwerde bei BAFA gegen VW/BMW/Audi eingereicht – mangelnde Transparenz bei menschenrechtlicher Sorgfalt in Lieferketten geltend gemacht ([ecchr.eu][4]).
* Audi reagierte via Business & Human Rights Resource Centre, bestätigt Code-of-Conduct und Mechanismen für Beschwerden, doch konkrete unabhängige Audit-Daten fehlen ([Business & Human Rights Resource Centre][6]).
* Mitarbeiterproteste in Audi-Brüssel reflektieren mögliche Missstände bei Löhnen und Mitbestimmung ([Reddit][7]).

#### **4. Datenschutz & Cybersicherheit**

* NGO-Berichte adressieren bisher primär menschenrechtliche Aspekte der Material- und Lieferketten; konkrete Bewertungen zum Datenschutz oder Datensparsamkeitsprinzip bei Audi existieren nicht öffentlich.

#### **5. Transparenz & LkSG-Reporting**

* Die EU Corporate Sustainability Due Diligence Directive (CSDDD) verschärft Reporting-Pflichten ab 2027, setzt aber auf bestehenden deutschen LkSG (in Kraft seit 1.1.2023) auf ([Financial Times][8]).
* Bisherige NGO- und Regierungsberichte nennen keine LkSG-Verstöße bei Audi/VW und attestieren stattdessen die Aufstellung von Compliance-Systemen und Lieferketten-Bewusstsein.
