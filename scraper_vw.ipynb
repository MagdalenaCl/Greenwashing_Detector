```python
import PyPDF2
import re
import nltk
from nltk.tokenize import sent_tokenize

# Stelle sicher, dass NLTK-Daten verfügbar sind
nltk.download('punkt')

# Schlüsselwörter für Nachhaltigkeit
SCHLUESSELWOERTER = [
    "klimaneutral", "co2-neutral", "nachhaltig", "umwelt", "klimaschutz",
    "recyceln", "erneuerbar", "emission", "energieeffizient",
    "verantwortung", "ökologisch", "grün", "CO2", "klimafreundlich", "ressourcenschonend"
]

import re

def satz_tokenizer(text):
    return re.split(r'(?<=[.!?]) +', text)


def pdf_zu_text(pfad_zur_pdf):
    text = ""
    with open(pfad_zur_pdf, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        for seite in reader.pages:
            seiten_text = seite.extract_text()
            if seiten_text:
                text += seiten_text + "\n"
    return text

def extrahiere_nachhaltigkeitsversprechen(text):
    saetze = satz_tokenizer(text)
    nachhaltigkeitssaetze = []

    for satz in saetze:
        for wort in SCHLUESSELWOERTER:
            if re.search(rf"\b{wort}\b", satz, re.IGNORECASE):
                nachhaltigkeitssaetze.append(satz.strip())
                break  # Nicht doppelt zählen, wenn mehrere Keywords

    return nachhaltigkeitssaetze

if __name__ == "__main__":
    pfad = "2023_Volkswagen_Group_Nachhaltigkeitsbericht.pdf"  # Pfad zur PDF-Datei
    print("PDF wird gelesen...")
    text = pdf_zu_text(pfad)

    print("Nachhaltigkeitsversprechen werden extrahiert...")
    versprechen = extrahiere_nachhaltigkeitsversprechen(text)

    if versprechen:
        print("\n--- Gefundene Nachhaltigkeitsversprechen ---\n")
        for idx, v in enumerate(versprechen, 1):
            print(f"{idx}. {v}")
    else:
        print("Keine Nachhaltigkeitsversprechen gefunden.")
```
